```python
from keras.models import Sequential 
from keras.layers import Dense 
from keras import Input
import pandas as pd 
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
data = pd.read_csv("Documents/kaggle competetion datasets /heart disease prediction /train.csv")
df = pd.DataFrame(data)
df.drop(["id"], axis = 1, inplace = True)
le = LabelEncoder()
ss = StandardScaler()
x = df.drop(["Heart Disease"], axis = 1)
x = pd.DataFrame(ss.fit_transform(x), columns = x.columns)
y = pd.DataFrame(le.fit_transform(df[["Heart Disease"]]))
#x.isna().sum()
y.head()
```

```python
test_data = pd.read_csv("Documents/kaggle competetion datasets /heart disease prediction /test.csv")
tst1 = pd.DataFrame(test_data)
tst1.drop(['id'], axis = 1, inplace = True)
tst = ss.fit_transform(tst1)
tst = pd.DataFrame(tst, columns = tst1.columns)


```

```python
col_nu = x.shape[1]
model = Sequential(
    [
        Input(shape = (col_nu,)),
        Dense(32, activation = 'relu'),
        Dense(64, activation = 'relu'),
        Dense(128, activation = 'relu'),
        Dense(1, activation = 'sigmoid')
    ]
)
model.compile(optimizer = 'adam', 
              loss = 'binary_crossentropy', 
              metrics = ['accuracy']
             )
model.fit(x, y, epochs = 15)
```

```python
ypred = (model.predict(tst) > 0.5).astype(int)
print(ypred)
```

```python
import numpy as np 
ypre = np.array(ypred)
ypre = ypre.flatten()
```

```python
prediction = pd.DataFrame({'id' : test_data['id'], 
                          "Heart Disease" : ypre})
```

```python
prediction.to_csv("submission_heart_1.csv", index = False)
```
